```python
job_date_cols = pd.DataFrame()
job_date_cols['job_date'] = job_date.columns
mb51_cols = pd.DataFrame()
mb51_cols['mb51'] = mb51.columns
dues_out_cols = pd.DataFrame()
dues_out_cols['dues_out'] = dues_out.columns
gr_price_cols = pd.DataFrame()
gr_price_cols['gr_price'] = gr_price.columns
soh_cols = pd.DataFrame()
soh_cols['soh'] = soh.columns
pr_cols = pd.DataFrame()
pr_cols['pr'] = pr.columns
```

```python
pd.concat([job_date_cols, mb51_cols, dues_out_cols, gr_price_cols, soh_cols, pr_cols], axis=1).to_csv('data_fields.csv',index=False)
```
